```python
import pandas as pd

# Load the dataset
df = pd.read_csv("chicago_crime_sample-3.csv")

# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Compute summary statistics for crime distribution across districts
crime_by_district = df['district'].value_counts()
district_mean = crime_by_district.mean()
district_std = crime_by_district.std()
district_min = crime_by_district.min()
district_max = crime_by_district.max()

# Compute overall arrest rate
arrest_rate = df['arrest'].mean() * 100

# Compute crime frequency by type
crime_type_counts = df['primary_type'].value_counts()

# Compute monthly crime trends
df['month'] = df['date'].dt.month
monthly_crime_counts = df.groupby('month').size()

# Compute arrest rate by crime type
arrest_rate_by_crime = df.groupby('primary_type')['arrest'].mean() * 100

# Display results
print("Mean crimes per district:", district_mean)
print("Standard deviation of crimes per district:", district_std)
print("Minimum crimes in a district:", district_min)
print("Maximum crimes in a district:", district_max)
print("Overall arrest rate (%):", arrest_rate)
print("Arrest rate by crime type:\n", arrest_rate_by_crime)
print("Monthly crime counts:\n", monthly_crime_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/chicago_crime_sample-3.csv")

# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month

def plot_crime_type_distribution(df):
    """Plots the distribution of different crime types."""
    plt.figure(figsize=(12,6))
    crime_counts = df['primary_type'].value_counts()
    sns.barplot(x=crime_counts.index, y=crime_counts.values)
    plt.xticks(rotation=90)
    plt.title("Crime Type Distribution")
    plt.xlabel("Crime Type")
    plt.ylabel("Number of Crimes")
    plt.show()


def plot_monthly_crime_trends(df):
    """Plots crime occurrences per month."""
    plt.figure(figsize=(10,5))
    monthly_counts = df.groupby('month').size()
    sns.lineplot(x=monthly_counts.index, y=monthly_counts.values, marker="o")
    plt.title("Monthly Crime Trends")
    plt.xlabel("Month")
    plt.ylabel("Number of Crimes")
    plt.show()


def plot_arrest_rate_by_crime(df):
    """Plots the arrest rate by crime type."""
    plt.figure(figsize=(12,6))
    arrest_rates = df.groupby('primary_type')['arrest'].mean() * 100
    sns.barplot(x=arrest_rates.index, y=arrest_rates.values)
    plt.xticks(rotation=90)
    plt.title("Arrest Rate by Crime Type")
    plt.xlabel("Crime Type")
    plt.ylabel("Arrest Rate (%)")
    plt.show()

# Generate the plots
plot_crime_type_distribution(df)
plot_monthly_crime_trends(df)
plot_arrest_rate_by_crime(df)
```
